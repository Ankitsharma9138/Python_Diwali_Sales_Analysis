# Exploratory Data Analysis

Gender

```python
sales.columns
```

```python
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(x = 'Gender',data=sales)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sales_gen =  sales.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data = sales_gen,x = 'Gender', y = 'Amount')
plt.show()
```

From above graphs we can see that most of the buyers are females and even the purchasing power of the females are greater than men.

# Age

```python
aa = sns.countplot(data = sales, x = 'Age Group',hue = 'Gender')
aa.bar_label(aa.containers[0])
plt.show()
```

```python
# Total Amount vs Age Group
sales_age = sales.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y = 'Amount')
plt.show()
```

From above graphs we can see that most of the buyers are of the age group between 26-35 years female

# State

```python
sales['State'].value_counts()
```

```python
# total numbers of orders from top 10 states
ax1 = sales.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(data=ax1,x='State',y='Orders',)
plt.xticks(rotation='vertical')
plt.show()
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and karnataka respectively.

# Marital Status

```python
ax3 = sns.countplot(data=sales,x='Marital_Status')
ax3.bar_label(ax3.containers[0])
plt.title('Marital Status')
plt.show()
```

```python
sales_status = sales.groupby(['Marital_Status','Gender'],as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_status,x='Marital_Status',y='Amount',hue='Gender')
plt.show()
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

# Occupation

```python
sns.set(rc={'figure.figsize':(20,7)})
ax5 = sns.countplot(data=sales,x='Occupation')
ax5.bar_label(ax5.containers[0])
plt.show()
```

```python
sales_occ = sales.groupby(['Occupation'], as_index=False).sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occ,x='Occupation',y='Amount')
plt.show()
```

From above graphs we can see that most of the buyers are working in IT Sector, HealthCare, Aviation.

# Product Category

```python
sales['Product_Category'].value_counts()
```

```python
ax6 = sns.countplot(data=sales,x='Product_Category')
ax6.bar_label(ax6.containers[0])
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Top 5
sales_pro = sales.groupby(['Product_Category'],as_index=False).sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(data=sales_pro,x='Product_Category',y='Amount')
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Footwear, Electronics category.

```python
sales_state = sales.groupby(['Product_ID'],as_index=False).sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_state,x='Product_ID',y='Orders')
plt.show()
```
